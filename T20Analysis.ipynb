```python
df=df.dropna()
```

```python
df.shape
```

```python
df['first innings score']=df['first innings score'].astype('int')
df['first innings wickets']=df['first innings wickets'].astype('int')
df['second innings score']=df['second innings score'].astype('int')
df['second innings wickets']=df['second innings wickets'].astype('int')
df['highest score']=df['highest score'].astype('int')
```

```python
df.dtypes
```

```python
grouped_data = df.groupby('winner')['winner'].count().sort_values(ascending=False)
plt.bar(grouped_data.index, grouped_data.values)
plt.xlabel("Team")
plt.ylabel("Number of matches won")
plt.title("Number of Matches Won by teams in t20 World Cup 2022")
plt.xticks(rotation=90)
plt.show()
```

```python
won_by=df['won by'].value_counts()
won_by
```

```python
won_by=df['won by'].value_counts()
label=won_by.index
counts=won_by.values
plt.pie(counts,labels=label,autopct='%1.2f%%')
plt.title("Number of Matches Won By Runs Or Wickets")
lst = []
for i in range(len(label)):
    lst.append(label[i] + " (" + str(counts[i]) + ")")
plt.legend(lst)
plt.show()
```

```python
toss=df['toss decision'].value_counts()
toss
```

```python
toss=df['toss decision'].value_counts()
label=toss.index
counts=toss.values
plt.pie(counts,labels=label,autopct='%1.2f%%')
plt.title("Toss Decision")
lst=[]
for i in range(len(label)):
    lst.append(label[i]+" ("+str(counts[i])+")")
plt.legend(lst)
plt.show()
```

```python
plt.bar(df['top scorer'],df['highest score'])
plt.xlabel('Player')
plt.ylabel('Highest Score')
plt.title('Top Scorers in t20 World Cup 2022')
plt.xticks(rotation=90)
plt.show()
```

```python
grouped_by=df.groupby('player of the match')["player of the match"].count().sort_values(ascending=False)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel("Player")
plt.ylabel("Number of awards won")
plt.title("Player of the Match Awards in t20 World Cup 2022")
plt.xticks(rotation=90)
plt.show()
```

```python
grouped_by=df.groupby('best bowler')['best bowler'].count().sort_values(ascending=False)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel('Player')
plt.ylabel('Number of awards won')
plt.xticks(rotation=90)
plt.show()
```

```python
x_labels = df['venue']
x = np.arange(len(x_labels)) 
width = 0.35
plt.bar(x - width/2, df['first innings score'], width)
plt.bar(x + width/2, df['second innings score'], width)
plt.xlabel("Stadium")
plt.ylabel('Score')
plt.title("Best Stadiums to Bat First or Chase")
plt.xticks(x, x_labels, rotation=90)
plt.legend(['first innings score','second innings score'])
plt.show()
```

```python
x_labels = df['venue']
x = np.arange(len(x_labels)) 
width = 0.35
plt.bar(x - width/2, df['first innings wickets'], width)
plt.bar(x + width/2, df['second innings wickets'], width)
plt.xlabel("Stadium")
plt.ylabel('Number of wickets')
plt.title("Best Stadiums to Bowl First or Defend")
plt.xticks(x, x_labels, rotation=90)
plt.legend(['first innings wickets','second innings wickets'])
plt.show()
```

```python
grouped_df = df.groupby('venue')[['first innings score','second innings score']].mean()
x_labels = grouped_df.index
x = np.arange(len(x_labels)) 
width = 0.35
plt.bar(x - width/2, grouped_df['first innings score'], width)
plt.bar(x + width/2, grouped_df['second innings score'], width)
plt.xlabel('Stadium')
plt.ylabel('Score')
plt.title("Best Stadiums to Bat First or Chase")
plt.xticks(x, x_labels, rotation=90)
plt.legend(['first innings score','second innings score'])
plt.show()
```

```python
grouped_df = df.groupby('venue')[['first innings wickets','second innings wickets']].mean()
x_labels = grouped_df.index
x = np.arange(len(x_labels)) 
width = 0.35
plt.bar(x - width/2, grouped_df['first innings wickets'], width)
plt.bar(x + width/2, grouped_df['second innings wickets'], width)
plt.xlabel('Stadium')
plt.ylabel('Number of wickets')
plt.title("Best Stadiums to Bowl First or Defend")
plt.xticks(x, x_labels, rotation=90)
plt.legend(['first innings wickets','second innings wickets'])
plt.show()
```
